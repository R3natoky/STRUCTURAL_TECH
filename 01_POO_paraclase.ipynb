### Otro ejemplo de Herencia

```python
class Material:

    def __init__(self, nombre, densidad):
        self.nombre = nombre
        self.densidad = densidad

    def calcular_peso(self, volumen):
        return volumen * self.densidad
    
class Acero(Material):
    
    def __init__(self, nombre, E=2000000, densidad=7850):
        self.E = E
        super().__init__(nombre, densidad)

class VarillaCorrugada(Acero):

    def __init__(self, nom, diam):
        super().__init__(nom)
        self.diametro = diam

    def calcular_area(self):    # Nuevo metodo
        return 3.14159265/4*(self.diametro)**2
    
class PerfilTubularCuadrado(Acero):

    def __init__(self, nombre, espesor, lado):
        super().__init__(nombre)
        self.espesor = espesor
        self.lado = lado

    def calcular_area(self):        # Sobreescribo método
        return 4 * self.lado * self.espesor
```

```python
tubo_100x100 = PerfilTubularCuadrado("100x100x3", 0.003, 0.100)
print(tubo_100x100.nombre)
print(tubo_100x100.densidad)
print(tubo_100x100.E)
print(tubo_100x100.espesor)
print(tubo_100x100.lado)
print(tubo_100x100.calcular_peso(1))
print(tubo_100x100.calcular_area())
```

## 7. Decoradores

### Ejemplo ilustrativo

```python
# Definiendo funcion: mi_funcion

def mi_funcion(num_1, num_2):
    suma = num_1 + num_2
    return suma
```

```python
mi_funcion(2,4)
```

```python
# Definiendo mi_decorador 

def mi_decorador(funcion_original):

    def funcion_modificada(num_1, num_2):
        print("Añadí esta impresión en pantalla")
        print(f"La suma es: {funcion_original(num_1, num_2)}")
        print("Fin de la decoración")
        return None
    
    return funcion_modificada

# Comentario: las funciones pueden definirse en cualquier
# lugar, incluso dentro de funciones
```

```python
# Decorando la funcion: mi_funcion
mi_funcion = mi_decorador(mi_funcion)
```

```python
mi_funcion(2, 4)
```

### Ejemplo usando sintaxis @

```python
# Equivalentemente

@mi_decorador
def mi_funcion_2(num_1, num_2):
    suma = num_1 + num_2
    return suma
```

```python
mi_funcion_2(2, 4)
```

```python
import time

# Definiendo funciones decoradoras
def no_hago_nada_nuevo(funcion_original):
    """Decorador que no hace nada"""

    def funcion_modificada(self, fuerza_axial):
        resultado = funcion_original(self, fuerza_axial)
        return resultado

    return funcion_modificada



# Definiendo funciones decoradoras
def validar_entrada(funcion_original):
    """Verifica que la fuerza sea menor a 10"""

    def funcion_modificada(self, fuerza_axial):
        if fuerza_axial > 10:
            raise ValueError("La fuerza axial debe ser menor a 10")
        resultado = funcion_original(self, fuerza_axial)
        return resultado

    return funcion_modificada



# Definiendo funciones decoradoras
def tiempo_ejecucion(funcion_original):
    """Mide e imprime el tiempo de ejecucion"""

    def funcion_modificada(self, fuerza_axial):
        inicio = time.time()
        resultado = funcion_original(self, fuerza_axial)
        fin = time.time()
        print(f"Demoró {fin-inicio} s")
        return resultado

    return funcion_modificada
```

```python
# Implementación de la clase BarraAxial
class BarraAxial:

    def __init__(self, area, modulo_elasticidad, longitud):
        self.area = area
        self.modulo_elasticidad = modulo_elasticidad
        self.longitud = longitud

    @tiempo_ejecucion
    @validar_entrada
    @no_hago_nada_nuevo
    def calcular_deformacion(self, fuerza_axial):
        time.sleep(0.3)    # Simulando un calculo costoso computacionalmente
        deformacion = fuerza_axial * self.longitud / (self.modulo_elasticidad * self.area)
        print(f"La deformación es: {deformacion}")
        return deformacion
    
    @staticmethod
    def calcular_suma_dos_numeros(num_1, num_2):
        suma = num_1 + num_2
        print(f"La suma de {num_1} y {num_2} es: {suma}")
        return suma
    
```

```python
# Instanciando objeto de clase BarraAxial
barra_de_armadura = BarraAxial(area=1, modulo_elasticidad=1, longitud=1)

# Usando el metodo calcular_deformacion
barra_de_armadura.calcular_deformacion(9)

# Usando funcionalidad de suma implementada en la clase
BarraAxial.calcular_suma_dos_numeros(3, 4)
```
